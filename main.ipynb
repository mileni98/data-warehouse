```python
import cx_Oracle
import pandas as pd

conStr = 'system/oracle@localhost:1521/xe'

conn = cx_Oracle.connect(conStr)

cur = conn.cursor()

sqlTxt = 'select * from "TEST1".country'
cur.execute(sqlTxt)

record = cur.fetchall()
print(record)

df= pd.DataFrame.from_records(record, columns = [x[0] for x in cur.description] )
print(df)
cur.close()
conn.close()
```
